```python
import copy

original = [[1, 2, 3], [4, 5, 6]]

shallow = copy.copy(original)

# Modify inner list
shallow[0].append(99)

print("Original:", original)
print("Shallow :", shallow)
```

```python
original.append([7,8,9])
original, shallow
```

```python
shallow.append([10,11,12])
original, shallow
```

```python
original[0][0] = 999
original, shallow
```

```python
shallow[0][1] = 9999
original, shallow
```

```python
import copy

original = [[1, 2, 3], [4, 5, 6]]

deep = copy.deepcopy(original)

# Modify inner list
deep[0].append(99)

print("Original:", original)
print("Deep    :", deep)
```

```python
original[0][0] = 999
original, deep
```

```python
deep[1][1] = 8888
original, deep
```

```python
original.append([10,11,12])
original, deep  
```
